## Function Definitions

Now we will define some functions that are used in the classification and data analysis section

```python
#colors for plots
death_palette=["darkcyan","tomato"]
one_color="darkcyan"

#perform classification (type determined by clf) on inputted training and test data
#clf: classifier object
#X_train: training data (unlabeled). used to fit classifier
#y_train: labels for training data. used to fit classifier
#X_test: test data (unlabeled). used test efficacy of fitted classifier with .predict
#y_test: labels for test data. used test efficacy of fitted classifier with .predict
#returns: name, accuracy, precision, recall, and fscore of clf as a list 
def calculate_scores(clf,X_train, y_train,X_test,y_test):
    #fit classifier to training data
    clf.fit(X_train, y_train)
    #for future results presentation
    name = clf.__class__.__name__
    if name=='Pipeline':name='SVM'
    train_predictions = clf.predict(X_test)
    # calculate accuracy, precision, recall, and fscore keeping first 4 decimals
    acc = float("{:.4}".format(accuracy_score(y_test, train_predictions)))
    precision = float("{:.4}".format(precision_score(y_test, train_predictions, average = 'macro'))) 
    recall =  float("{:.4}".format(recall_score(y_test, train_predictions, average = 'macro')))
    f_score = float("{:.4}".format(f1_score(y_test, train_predictions, average = 'macro')))
    return [name,acc,precision,recall,f_score]

#Clean dataframe column name formatting for better graphical presentation. 
#Edits df directly so nothing is returned
#df: dataframe to clean
def clean_names(df):
    column_names=[]
    for name in df.columns:
        if name.isupper():
            name=name[0]+name[1:].lower().replace("_"," ")
        else:
            name=name[0].upper()+name[1:].replace("_"," ")
        column_names.append(name)
    df.columns=column_names
    return
#Creates a seaborn violin plot and edits it for better presentation
#df: dataframe we will pull the information on 
#col: event to compare with death event
#axs (optional): where figure should be on a multigrid, default is that the graph is not on a grid
#title_string: title of plot
def clean_violin(df,col,axs=None,title_string=""):
    p=sns.violinplot(y=df[col],x=df["Death event"],palette=death_palette,scale="area",ax=axs,linewidth=2) 
    p.set(xticklabels=["Living","Deceased"],xlabel="State at follow-up")
    sns.despine(left=True,bottom=True)
    return p
#Creates a seaborn count plot and edits it for better presentation
#df: dataframe we will pull the information on 
#col: event to compare with death event
#axs (optional): where figure should be on a multigrid, default is that the graph is not on a grid
#title_string: title of plot
def clean_bar(df,col,axs=None,title_string=""):
    p=sns.countplot(x=df[col],hue=df["Death event"],palette=death_palette,ax=axs)        
    if col!="Sex":p.set(xticklabels=["No","Yes"])
    else:p.set(xticklabels=["Female","Male"])
    p.legend(labels=["Living","Deceased"],title="State at follow-up")

    sns.despine(left=True,bottom=True)
    return p
    
#Creates a seaborn kde plot and edits it for better presentation
#df: dataframe we will pull the information on 
#col: event to compare with death event
#axs (optional): where figure should be on a multigrid, default is that the graph is not on a grid
#title_string: title of plot
def clean_kde(df,col,axs=None,title_string=""):
    p=sns.kdeplot(data=df, x=col, hue="Death event", palette=death_palette,alpha=0.7,multiple="fill",ax=axs)
    p.legend(labels=["Deceased","Living"],title="State at follow-up")
    sns.despine(left=True,bottom=True)
    return p

#Creates a seaborn subplot of kde figures and edits it for better presentation
#df: dataframe we will pull the information on 
#cols_to_plot: list of columns that will be pulled from df and presented as a kde plot 
#num_cols: number of columns on the grid
#num_rows: number of rows on the grid
def plot_grid_kde(df,cols_to_plot,num_cols,num_rows):
    fig, axs = plt.subplots(ncols=num_cols,nrows=num_rows,sharex=False)
    i=0
    j=0
    for d in cols_to_plot:
        if j<num_cols:
            clean_kde(df,d,axs[i,j],title_string="")        
        else:
            i+=1
            j=0
            clean_kde(df,d,axs[i,j],title_string="")
        j+=1 
    fig.tight_layout(pad=3.0)
    
    
#Creates a seaborn subplot of violin figures and edits it for better presentation
#df: dataframe we will pull the information on 
#cols_to_plot: list of columns that will be pulled from df and presented as a kde plot 
#num_cols: number of columns on the grid
#num_rows: number of rows on the grid
def plot_grid_disc(df,cols_to_plot,num_cols,num_rows):
    fig, axs = plt.subplots(ncols=num_cols,nrows=num_rows,sharex=False)
    i=0
    j=0
    for d in cols_to_plot:
        if j<num_cols:
            clean_violin(df,d,axs[i,j],title_string="")        
        else:
            i+=1
            j=0
            clean_violin(df,d,axs[i,j],title_string="")
        j+=1 
    fig.tight_layout(pad=3.0)
    
#Creates a seaborn subplot and edits it for better presentation
#df: dataframe we will pull the information on 
#cols_to_plot: list of columns that will be pulled from df and presented as a kde plot 
#num_cols: number of columns on the grid
#num_rows: number of rows on the grid    
def plot_grid_bin(df,cols_to_plot,num_cols,num_rows):
    fig, axs = plt.subplots(ncols=num_cols,nrows=num_rows,sharex=False)
    i=0
    j=0
    for d in cols_to_plot:
        if j<num_cols:
            clean_bar(df,d,axs[i,j],title_string="") 
            
        else:
            i+=1
            j=0
            clean_bar(df,d,axs[i,j],title_string="")
        j+=1 
    fig.tight_layout(pad=3.0)
```

### Data explanation:


```python
df=pd.read_csv("heart_failure_clinical_records_dataset.csv")
df['age'] = df['age'].astype(int)
column_names=[]
clean_names(df)
df.head()
```

```python
print(df.shape)
```

We have 299 patients and 13 features in the dataset

Meaning of the 13 features (columns)
- Age: Age of the patient (years)
- Anaemia: Decrease of red blood cells or hemoglobin (boolean: 0=false, 1=true)
- High blood pressure: If a patient has hypertension (boolean: 0=false, 1=true)
- Creatinine phosphokinase: Level of the CPK enzyme in the blood (mcg/L)
- Diabetes: If the patient has diabetes (boolean: 0=false, 1=true)
- Ejection fraction: Percentage of blood leaving the heart at each contraction (percentage)
- Platelets: Platelets in the blood (kiloplatelets/mL)
- Sex: Woman or man (binary: 0=woman, 1=man)
- Serum creatinine: level of serum creatinine in the blood (mg/dL)
- Serum sodium: level of serum sodium in the blood (mEq/L)
- Smoking: if the patient smokes or not (boolean: 0=false, 1=true)
- Time: follow-up period (days)
- Death event: if the patient is deceased during the follow-up period (boolean: 0=false, 1=true)

Source:https://bmcmedinformdecismak.biomedcentral.com/articles/10.1186/s12911-020-1023-5

## Classification

Now that we have a better understand the data we are working with, lets move on to classification. We will first split the dataframe into training and test data. 

```python
X=df.drop(["Death event"],axis=1)
y = df["Death event"]
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=32)
```

Now we have split the train and test indices while ensuring a .2 .8 split. Next we should ensure that the number of living and dead patients is even across the training and test data

```python
print("Training data split\n",y_train.value_counts(normalize=True))
print("\nTest data split\n",y_test.value_counts(normalize=True))
```

The data has been split evenly. Now we will begin fine tuning our three classifiers with gridsearchcv. We will be using random forest, KNN, and linear discriminant analysis.

```python
#fine tuning randomforest:
paramfeat = {'max_features':['sqrt', 'log2'], 'criterion': ['gini', 'entropy']}
random_forest = RandomForestClassifier(n_estimators=1000)
search = GridSearchCV(random_forest, paramfeat, cv=3)
search.fit(X_train, y_train)
feats = search.best_params_.get('max_features') 
cr = search.best_params_.get('criterion')
random_forest = RandomForestClassifier(max_features = feats, criterion = cr)
```

```python
print("We will be using", feats, "and", cr, "with random forest")
```

```python
#fine tuning k-nn
k_nn = KNeighborsClassifier()
ks = {'n_neighbors': np.arange(1, 20)}
k_nn = GridSearchCV(k_nn, ks, cv = 3)
k_nn.fit(X_train, y_train)
neighbors=k_nn.best_params_['n_neighbors']
k_nn = KNeighborsClassifier(n_neighbors = neighbors)
```

```python
print("We will have", neighbors, "neighbors with KNN")
```

```python
#fine tuning linear discriminant analysis
solver={'solver':['svd', 'lsqr', 'eigen']}
LDA=LinearDiscriminantAnalysis()
#choosing solver based on acc
search = GridSearchCV(LDA, solver, scoring='accuracy')
results = search.fit(X_train, y_train)
LDA=LinearDiscriminantAnalysis(solver=results.best_params_['solver'])
```

```python
print("We will use the", results.best_params_['solver'], "solver with LDA")
```

Now that our classifiers best suit our data we can actually begin to perform classification and look at some resulting metrics to see which classifier has performed the best.

```python
res=[]
for clf in [random_forest,k_nn,LDA]:
    name=clf.__class__.__name__
    results=calculate_scores(clf,X_train, y_train,X_test,y_test)
    res.append(calculate_scores(clf,X_train, y_train,X_test,y_test))
classification_res=pd.DataFrame(res,columns=["Name",'Accuracy','Precision','Recall','Fscore'])
```

```python
classification_res
```

Random Forest and LDA performed fairly similarly on all fronts and KNN is a distant third in performace. Now lets examine the confusion matrices of these three classifiers

```python

matrices=[]
scores=[]
fig, axs = plt.subplots(nrows=3,sharey=True)
i=0
for clf in [random_forest,k_nn,LDA]:
    name=clf.__class__.__name__
    predictions = clf.predict(X_test)
    score = clf.score(X_test, y_test)
    cm = metrics.confusion_matrix(y_test, predictions,normalize="true")
    matrices.append(cm)
    scores.append(score)
    p=sns.heatmap(cm, annot=True, linewidths=1, cmap = 'Reds',ax=axs[i]);
    i+=1
    p.set(title=name+" confusion matrix",xlabel="Predicted label",ylabel="Actual label",xticklabels=["Living","Deceased"],yticklabels=["Living","Deceased"])
fig.tight_layout(pad=3.0)
    
```

We can now see why KNN performs so poorly; in its current state it always predicts the patient survives up to the follow-up period. It is able to predict above chance simply because more than 50% of the patients in the dataset ended up surviving to the follow-up period
